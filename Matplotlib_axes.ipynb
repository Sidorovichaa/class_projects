```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('students.csv')
df
```

```python
fig = plt.figure()
ax00 = fig.add_subplot(331)
ax00.hist(df['height'], color = 'firebrick')
ax00.set(xlabel='height', ylabel='students')
ax00.grid(True, color = 'w', linewidth = 2)
ax00.set_axisbelow(True)
ax00.set_facecolor('lightgray')
arrow1 = plt.Arrow(167, 1500, 0, 300, color = 'black', lw = 2)
ax00.add_artist(arrow1)

ax10 = fig.add_subplot(334)
ax10.scatter(df['height'], df['weight'], color = 'firebrick')
ax10.grid(True, color = 'w', linewidth = 2)
ax10.set_axisbelow(True)
ax10.set(xlabel='height', ylabel='weight')
ax10.set_facecolor('lightgray')
ax10.plot(np.mean(df.height), np.mean(df.weight), marker='x', c = 'black')

ax11 = fig.add_subplot(335)
ax11.hist(df['weight'], color = 'firebrick')
ax11.set(xlabel='weight', ylabel='students')
ax11.grid(True, color = 'w', linewidth = 2)
ax11.set_axisbelow(True)
ax11.set_facecolor('lightgray')
arrow2 = plt.Arrow(68, 2050, 0, 410, color = 'black', lw = 2)
ax11.add_artist(arrow2)

ax20 = fig.add_subplot(337)
ax20.scatter(df['height'], df['salary'], color = 'firebrick')
ax20.grid(True, color = 'w', linewidth = 2)
ax20.set_axisbelow(True)
ax20.set(xlabel='height', ylabel='salary')
ax20.set_facecolor('lightgray')
ax20.plot(np.mean(df.height), np.mean(df.salary), marker='x', c = 'black')

ax21 = fig.add_subplot(338)
ax21.scatter(df['weight'], df['salary'], color = 'firebrick')
ax21.grid(True, color = 'w', linewidth = 2)
ax21.set_axisbelow(True)
ax21.set(xlabel='weight', ylabel='salary')
ax21.set_facecolor('lightgray')
ax21.plot(np.mean(df.weight), np.mean(df.salary), marker='x', c = 'black')

ax22 = fig.add_subplot(339)
ax22.hist(df['salary'], color = 'firebrick')
ax22.set(xlabel='salary', ylabel='students')
ax22.grid(True, color = 'w', linewidth = 2)
ax22.set_axisbelow(True)
ax22.set_facecolor('lightgray')
arrow3 = plt.Arrow(40000, 350, 0, 100, color = 'black', lw = 2)
ax22.add_artist(arrow3)

fig.set_size_inches(20,13)
plt.savefig('HW11apr.png')

plt.show()
```
