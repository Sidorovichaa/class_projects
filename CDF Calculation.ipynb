```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import json
data = json.load(open('data.json'))
dat = data['result'][0]['data']
df = pd.read_json(json.dumps(dat), orient='records')
```

```python
df.head()
```

```python
stud_id = df['stud.id'].sort_values()
```

```python
stud_id_cump = []
for i, x in enumerate(stud_id, 1):
    stud_id_cump.append((i-0.5)/len(stud_id))
```

```python
sns.scatterplot(x = stud_id, y = stud_id_cump)
plt.ylabel('CDF')
plt.xlabel('stud.id')
```

```python
height = df['height'].sort_values()
height_cump = []
for i, x in enumerate(height, 1):
    height_cump.append((i-0.5)/len(height))
sns.scatterplot(x = height, y = height_cump)
plt.ylabel('CDF')
plt.xlabel('height')
```

```python
salary = df['salary'].sort_values()
salary_cump = []
for i, x in enumerate(salary, 1):
    salary_cump.append((i-0.5)/len(salary))
sns.scatterplot(x = salary, y = salary_cump)
plt.ylabel('CDF')
plt.xlabel('salary')
```

```python
age = df['age'].sort_values()
age_cump = []
for i, x in enumerate(age, 1):
    age_cump.append((i-0.5)/len(age))
sns.scatterplot(x = age, y = age_cump)
plt.ylabel('CDF')
plt.xlabel('age')
```

```python
fig, axs = plt.subplots(1,4, sharey = True)
sns.scatterplot(x = stud_id, y = stud_id_cump, ax = axs[0])
plt.ylabel('CDF')
plt.xlabel('stud.id')

sns.scatterplot(x = height, y = height_cump, ax = axs[1])
plt.xlabel('height')

sns.scatterplot(x = salary, y = salary_cump, ax = axs[2])
plt.xlabel('salary')

sns.scatterplot(x = age, y = age_cump, ax = axs[3])
plt.xlabel('age')

fig.set_size_inches(20,5)
```

```python
norm_stud_id = (stud_id - stud_id.mean())/stud_id.std()
```

```python
norm_height = (height - height.mean())/height.std()
```

```python
norm_salary = (salary - salary.mean())/salary.std()
```

```python
norm_age = (age - age.mean())/age.std()
```

```python
fig, axs = plt.subplots(4,1, sharey = False)
sns.scatterplot(x = norm_stud_id, y = stud_id_cump, ax = axs[0])
plt.xlabel('norm_stud_id')

sns.scatterplot(x = norm_height, y = height_cump, ax = axs[1])
plt.xlabel('norm_height')

sns.scatterplot(x = norm_salary, y = salary_cump, ax = axs[2])
plt.xlabel('norm_salary')

sns.scatterplot(x = norm_age, y = age_cump, ax = axs[3])
plt.xlabel('norm_age')
axs[3].set_xlim (-8, 8)


fig.set_size_inches(3,15)
```
