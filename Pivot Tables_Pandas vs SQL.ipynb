```python
%%time 
data = json.load(open('data.json'))
dat = data['result'][0]['data']
df = pd.read_json(json.dumps(dat), orient='records')
pv_table = pd.pivot_table(df,index=["religion"],values=["salary", "age", "weight", "graduated"],
               aggfunc={"salary":[len, np.mean], "age": len, "weight": [np.max, np.min], "graduated": np.sum},fill_value=0)
pv_table
```

```python
%%time
connector = sqlite3.connect('studinfo3.db')
cursor = connector.cursor()
df.to_sql('studinfo', connector)
sql3 = """
SELECT religion AS religion,
       count(salary) AS "COUNT(salary)",
       count(DISTINCT age) AS "COUNT_DISTINCT(age)",
       AVG(salary) AS "AVG(salary)",
       max(weight) AS "MAX(weight)",
       min(weight) AS "MIN(weight)",
       sum(graduated) AS "SUM(graduated)"
FROM studinfo
GROUP BY religion
"""
cursor.execute(sql3)
for row in cursor:
    print(row)
cursor.execute(sql3)
df3 = pd.read_sql(sql3, connector)
print(df3)
```
