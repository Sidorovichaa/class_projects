```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
mu = 10
sigma = 2
```

```python
n = 50
```

```python
#дисперсия всей популяции известна sigma
tscore = []
for i in range(10000):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    z = (sample.mean()-mu)/(sigma/pow(n, 0.5))
    tscore.append(z)
```

```python
#дисперсия всей популяции известна sigma
tscore = []
n = 2
for i in range(10000):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    z = (sample.mean()-mu)/(sigma/pow(n, 0.5))
    tscore.append(z)
```

```python
#дисперсия всей популяции не известна 
#малая выборка
n = 3
tscore = []
for i in range(10000):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    s = sample.std()
    z = (sample.mean()-mu)/(s/pow(n, 0.5))
    tscore.append(z)
```

```python
#дисперсия всей популяции не известна 
#большая выборка > 30
n = 40
tscore = []
for i in range(10000):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    s = sample.std()
    z = (sample.mean()-mu)/(s/pow(n, 0.5))
    tscore.append(z)
```

```python
tscore = np.array(tscore)
m = tscore.mean()
xx = np.linspace(-5, 5, 100)
sns.histplot(data=tscore, stat='density')
yy = norm.pdf(xx)
sns.lineplot(x=xx, y=yy, color='k')
plt.xlim(-5, 5)
plt.xlabel(r'$t=\frac{\bar{v}-\mu}{std(v)/\sqrt{n}}$')
```
